```python
class Report:
    bestCombination     : list
    notUsedItems        : list
    bestCombinationValue: int
    notUsedItemsValue   : int

    def __init__ (self, bestCombinationI, notUsedItemsI, 
                  bestCombinationValueI, notUsedItemsValueI):
        self.bestCombination      = bestCombinationI
        self.notUsedItems         = notUsedItemsI
        self.bestCombinationValue = bestCombinationValueI
        self.notUsedItemsValue    = notUsedItemsValueI
    
    def getGeneralReport(self):
        Answer = "La mejor combinación de objetos según los elementos dados y el paso máximo de la maleta son: "

        for i in range(len(self.bestCombination)):
            Answer = Answer + str(self.bestCombination[i].getComboWV())
        
        Answer = Answer + "\n" + "Los objetos que no se llevan son: "

        for i in range(len(self.notUsedItems)):
            Answer = Answer + str(self.notUsedItems[i].getComboWV())

        Answer = Answer + "\n" + "El valor de los objetos llevados es de: " + str(self.bestCombinationValue) + "\n" +  "Y el valor de los objetos no llevados es de: " + str(self.notUsedItemsValue) + "\n"

        print(Answer)
        return 
    
class Items:
    comboWV : list
    weight  : int
    value   : int

    def __init__(self, weightI, valueI):
        self.comboWV = [weightI, valueI]
        self.weight = weightI
        self.value = valueI

    def getComboWV(self):
        return self.comboWV

    def getWeight(self):
        return self.weight
    
    def getValue(self):
        return self.value

class backPack:
    maximumW     : int
    items        : list
    report       : Report
    
    def __init__ (self, maximumWI, itemsI):
        self.maximumW = maximumWI
        self.items    = itemsI
        self.report   = self.calculateBestCombination(itemsI)

    def getReport(self):
        return self.report

    def calculateBestCombination(self, listItems):

        allCombinations = []
        allCombinations.append(listItems)

        for i in range(len(listItems)):
            auxiliarList = listItems.copy()
            auxiliar = listItems[i]
            auxiliarList.remove(auxiliar)
            auxiliarList.append(auxiliar)

            for j in range(len(listItems)):
                for k in range(j+1, len(listItems)):
                    auxiliarList2 = auxiliarList.copy()
                    auxiliarList3 = auxiliarList.copy()
                    for r in range(len(auxiliarList[j:k])):
                        auxiliarList2.remove(auxiliarList3[j:k][r])
                    allCombinations.append(auxiliarList2)

        weightAcum = 0
        valueAcum = 0
        pastValue = 0
        position = 0

        for i in range(len(allCombinations)):
            for j in range(len(allCombinations[i])):
                weightAcum = weightAcum + allCombinations[i][j].getWeight()
                valueAcum = valueAcum + allCombinations[i][j].getValue()
            if weightAcum <= self.maximumW and valueAcum > pastValue:
                pastValue = valueAcum
                position = i
            valueAcum = 0
            weightAcum = 0

        notUsedItems = listItems.copy()
        for i in range(len(allCombinations[position])):
            notUsedItems.remove(allCombinations[position][i])

        valueAcum2 = 0

        for i in range(len(notUsedItems)):
            valueAcum2 = valueAcum2 + notUsedItems[i].getValue()

        report1 = Report(allCombinations[position], notUsedItems, pastValue, valueAcum2)

        return report1
    
if __name__ == "__main__":

    maximumWeight = 10
    listOfItems = [Items(2,3), Items(3,4), Items(4,5), Items(5,6)]
    back1 = backPack(maximumWeight, listOfItems)
    back1.getReport().getGeneralReport()
```
